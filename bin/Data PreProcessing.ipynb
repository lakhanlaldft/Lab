```python
import pandas as pd
import os
```

This notebook converts the JSON file downloaded from Citrination to a simpler csv format to be used throughout the lab.
https://citrination.com/datasets/1160/show_files

```python
df = pd.read_json("../data/band_gap_jpcrd_transformed-pif-onlyRef.json")
```

```python
prop_df = pd.DataFrame()
for idx in df.index:
    prop_list = df["properties"][idx]
    column_list = list()
    prop_detail_list = list()
    for prop in prop_list:
        prop_name = prop["name"]
        prop_value = prop["scalars"][0]["value"]
        column_list.append(prop_name+" value")
        prop_detail_list.append(prop_value)
#         print(prop_list)
        try:
            prop_units = prop["units"]
            column_list.append(prop_name+" units")
            prop_detail_list.append(prop_units)
            
            prop_conditions = prop["conditions"]
            column_list.append(prop_name+" conditions")
            prop_detail_list.append(prop_conditions)
            
            prop_method = prop["method"]["name"]
            column_list.append(prop_name+" method")
            prop_detail_list.append(prop_method)
            
            if len(prop["method"])>1:
                print("warning method length > 1 for index",idx)
            if len(prop["scalars"])>1:
                print("warning scalars length > 1 for index",idx)
        except:
            pass
#     print(prop_list)  
    

    temp_prop_df = pd.DataFrame([prop_detail_list],columns=column_list)
    try:
        temp_prop_df["Phase values"] = temp_prop_df["Phase value"].values.tolist()
        temp_prop_df = temp_prop_df.drop(["Phase value"], axis=1)
#         temp_prop_dftemp_prop_df.drop(["Phase value"], axis=1)
    except:
        pass
    
    prop_df = prop_df.append(temp_prop_df,sort=False,ignore_index=True)
```

```python
df = df.merge(prop_df,how="outer",left_index=True,right_index=True)
```

```python
# merge in reliability
df_merge = pd.read_csv("../data/data_merge_clean.csv")

df = df.merge(df_merge,how="outer",left_index=True,right_index=True)
```

```python
# output to csv for use in MASTML
mastml_df = pd.DataFrame([df["chemicalFormula Clean"],df["Band gap values Clean"],df["Band gap units"],df["Band gap method"],df["Reliability"]]).T
```

```python
# drop missing bandgap values
mastml_df = mastml_df[mastml_df["Band gap values Clean"].notna()]
mastml_df.index.rename("index",inplace=True)
```

```python
# Output data to csv - Include above as seperate notebook, but not included in lab
output_path = "../data/bandgap_data_v2.csv"

if os.path.isfile(output_path):
    print(output_path," exists, not creating new file")
else:
    mastml_df.to_csv(output_path)
```
